# The Sierpinski Triangle

```python
# Import necessary libraries
from numpy.random import random, randint
import matplotlib.pyplot as plt
```

```python
# Constants
plt_max_x = 1000.0
plt_max_y = 1000.0
iterations = 24999
```

```python
# Choose 3 points to form an equilateral triangle
# where 0 < x < plt_max_x
# and 0 < y < plt_max_y

anchor_points = [(100.0, 100.0), (900.0, 100.0), (500.0, 792.82)]
```

```python
# Empty list to store our points
lst_points = []
```

```python
# Choose a random point that (hopefully) falls into this equilateral triangle
starting_point = tuple(random(2) * plt_max_x)
starting_point
```

```python
# To gurantee that it does fall in the triangle, apply the Sierpinski Triangle rule once
next_point = tuple((
      anchor_points[0][0] + (starting_point[0] - anchor_points[0][0]) / 2.0
    , anchor_points[0][1] + (starting_point[1] - anchor_points[0][1]) / 2.0
))

# Add this starting point to our list of points
lst_points.append(next_point)
lst_points
```

```python
# Iterate through enough times to create the Sierpinski Triangle, adding each new point to the list
for _ in range(iterations):
    i = randint(3)
    last_point = lst_points[-1]
    next_point = tuple((
          anchor_points[i][0] + (last_point[0] - anchor_points[i][0]) / 2.0
        , anchor_points[i][1] + (last_point[1] - anchor_points[i][1]) / 2.0
    ))
    lst_points.append(next_point)
```

```python
len(lst_points)
```

```python
# From the generated list of points, extract the x and y coordinates into separate lists for scatter plotting
lst_x = [n[0] for n in lst_points]
lst_y = [n[1] for n in lst_points]
```

```python
# Plot the triangle
plt.figure(figsize=(10,10))
plt.scatter(lst_x, lst_y, s=1)
plt.savefig('SierpinskiTriangle.png')
```
